## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from newsapi import NewsApiClient
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
api_key = os.getenv("NEWS_API_KEY")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
btc_headlines = newsapi.get_everything(
    q="bitcoin",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Bitcoin: {btc_headlines['totalResults']}")

# Show sample article
btc_headlines["articles"][0]
```

```python
# Fetch the Ethereum news articles
eth_headlines = newsapi.get_everything(
    q="ethereum",
    language="en",
    page_size=100,
    sort_by="relevancy"
)

# Print total articles
print(f"Total articles about Bitcoin: {eth_headlines['totalResults']}")

# Show sample article
eth_headlines["articles"][0]
```

```python
# Create the Bitcoin sentiment scores DataFrame
btc_sentiments = []

for article in btc_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        btc_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
btc_df = pd.DataFrame(btc_sentiments)

# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
btc_df = btc_df[cols]

btc_df.head()
```

```python
# Create the Ethereum sentiment scores DataFrame
eth_sentiments = []

for article in eth_headlines["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        eth_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
# Create DataFrame
eth_df = pd.DataFrame(eth_sentiments)

# Reorder DataFrame columns
cols = ["date", "text", "compound", "positive", "negative", "neutral"]
eth_df = eth_df[cols]

eth_df.head()
```

```python
# Describe the Bitcoin Sentiment
btc_df.describe()
```

```python
# Describe the Ethereum Sentiment
eth_df.describe()
```

Q: Which coin had the highest mean positive score?

A: Ethereum

Q: Which coin had the highest compound score?

A: Ethereum

Q. Which coin had the highest positive score?

A: Ethereum

## 2. Natural Language Processing
---
###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
sw = set(stopwords.words('english'))

# Expand the default stopwords list if necessary
sw_addons = {'said', 'mln', 'chars','reuters'}
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the punctuation from text
    # Define the regex parameters
    regex = re.compile("[^a-zA-Z ]")

    # Apply regex parameters to article
    re_clean = regex.sub('', text)
   
    # Create a tokenized list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
   
    # Convert the words to lowercase and Remove the stop words
    tokens =[word.lower() for word in words if word.lower() not in sw.union(sw_addons)]
    
    return tokens
```

```python
# Create a new tokens column for Bitcoin
t = [tokenizer(text) for text in btc_df['text']]
btc_df['tokens'] = t
btc_df.head()
```

```python
# Create a new tokens column for Ethereum
t = [tokenizer(text) for text in eth_df['text']]
eth_df['tokens'] = t
eth_df.head()
```

### NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
processed = []
# Summarize all articles' tokens
for tokens in btc_df['tokens']:
    for token in tokens:
        processed.append(token)

btc_bigram = Counter(ngrams(processed, n=2))
btc_processed = processed
```

```python
# Generate the Ethereum N-grams where N=2
processed = []
# Summarize all articles' tokens
for tokens in eth_df['tokens']:
    for token in tokens:
        processed.append(token)

eth_bigram = Counter(ngrams(processed, n=2))
eth_processed = processed
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=3):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin
print(dict(btc_bigram.most_common(10)))
```

```python
# Use token_count to get the top 10 words for Ethereum
print(dict(eth_bigram.most_common(10)))
```

### Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
# Generate the Bitcoin word cloud
w = Counter(btc_processed)
#word_df.head()
btc_word_df = pd.DataFrame.from_dict(w, orient='index').reset_index()
btc_word_df.rename(columns={'index':'word', 0:'count'}, inplace=True)


# Order the DataFrame by word frequency in descending order
btc_word_df = btc_word_df.sort_values(by=["count"], ascending=False)
btc_word_df.head()
```

```python
# Create a string list of terms to generate the word cloud
terms_list = str(btc_word_df["word"].tolist())

wordcloud = WordCloud(colormap="RdYlBu").generate(terms_list)
plt.imshow(wordcloud)
plt.axis("off")
fontdict = {"fontsize": 20, "fontweight": "bold"}
plt.title("Money News Word Cloud", fontdict=fontdict)
plt.show()
```

```python
# Generate the Ethereum word cloud
w = Counter(eth_processed)
#word_df.head()
eth_word_df = pd.DataFrame.from_dict(w, orient='index').reset_index()
eth_word_df.rename(columns={'index':'word', 0:'count'}, inplace=True)


# Order the DataFrame by word frequency in descending order
eth_word_df = eth_word_df.sort_values(by=["count"], ascending=False)
eth_word_df.head()
```

```python
# Create a string list of terms to generate the word cloud
terms_list = str(eth_word_df["word"].tolist())

wordcloud = WordCloud(colormap="RdYlBu").generate(terms_list)
plt.imshow(wordcloud)
plt.axis("off")
fontdict = {"fontsize": 20, "fontweight": "bold"}
plt.title("Money News Word Cloud", fontdict=fontdict)
plt.show()
```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
#!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
btc_text = ''
for i in btc_df['text']:
    btc_text = btc_text + i
```

```python
# Run the NER processor on all of the text
bdoc = nlp(btc_text)

# Add a title to the document
bdoc.user_data["title"] = "Bitcoin NER"
```

```python
# Render the visualization
displacy.render(bdoc, style='ent')
```

```python
# List all Entities
print('The entitites are:')

labels = []

for ent in bdoc.ents:
    if ent.label_ not in labels:
        labels.append(ent.label_)

print([l for l in labels])
```

---

### Ethereum NER

```python
# Concatenate all of the Ethereum text together
eth_text = ''
for i in eth_df['text']:
    eth_text = eth_text + i
```

```python
# Run the NER processor on all of the text
edoc = nlp(eth_text)

# Add a title to the document
edoc.user_data["title"] = "Ethereum NER"
```

```python
# Render the visualization
displacy.render(edoc, style='ent')
```

```python
# List all Entities
print('The entitites are:')

labels = []

for ent in edoc.ents:
    if ent.label_ not in labels:
        labels.append(ent.label_)

print([l for l in labels])
```
